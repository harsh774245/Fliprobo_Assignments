```python
text=re.sub(r'\[[0-9]*\]',' ',text)
text=re.sub(r'\s+',' ',text)
text = re.sub(r'\d',' ',text)
text=text.replace(',','')
text=text.lower()
```

```python
sentences = nltk.sent_tokenize(text)
sentences
```

```python
sent_word = [nltk.word_tokenize(sentence) for sentence in sentences]
sent_word
```

```python
stopword=stopwords.words('english')
stopword.append('.')
```

```python
for i in range(len(sent_word)):
    sent_word[i] = [word for word in sent_word[i]
                    if word not in stopword]
```

```python
sent_word
```

```python
model=Word2Vec(sent_word,min_count=1)
words=model.wv.vocab
```

```python
len(words)
```

```python
vector=model.wv['swami']
print(vector)
```

```python
word_list=['swami','study','bed','offended','homework','mathematics',
                            'closer','play','headmaster','fight','surprise','friends','home',
                            'mother','exam','quickly','classmates','promise','enemies','except',
                            'faster','write','efforts']
for word in word_list:
    sim = model.wv.most_similar(word,topn=5)
    print(word)
    print(sim)
    print('\n\n')
```

```python
from sklearn.manifold import TSNE
word_list = ['swami','study','bed','offended','homework','mathematics',
                            'closer','play','headmaster','fight','surprise','friends','home',
                            'mother','exam','quickly','classmates','promise','enemies','except',
                            'faster','write','efforts']
def tsne(model):
    labels = []
    wordvecs = []

    for word in word_list:
        wordvecs.append(model[word])
        labels.append(word)
    
    tsne_model = TSNE(perplexity=4, n_components=2, init='pca', random_state=42)
    coordinates = tsne_model.fit_transform(wordvecs)

    x = []
    y = []
    for value in coordinates:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(12,8)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(2, 2),
                     textcoords='offset points', ha='left')

    plt.show()

tsne(model)
```
