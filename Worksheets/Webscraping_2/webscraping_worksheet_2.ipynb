```python
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import requests
import urllib.request
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

```python
browser = webdriver.Firefox()
url = browser.get('https://www.flipkart.com/search?q=rayban')
images=browser.find_elements_by_xpath("//img[@class='_3togXc']")
```

```python
images=images[:10]
```

```python
len(images)
```

```python
product_data={}
product_data['image_urls']=[]
for image in images:
    source=image.get_attribute('src')
    product_data['image_urls'].append(source)
```

```python
import shutil
import os
import requests
import pandas as pd
import csv

current_path=os.getcwd()
path=os.path.join(current_path,'rayban')
if not os.path.exists(path):
    os.makedirs(path)

for index, link in enumerate(product_data['image_urls']):
            
        print('Downloading {0} image'.format(index+1))
        response = requests.get(link)
        with open('{0}/img_{1}.jpg'.format('rayban',index), 'wb') as file:
            file.write(response.content)
```

## Method 3

```python
# Entire code in a single loop
```

```python
href=[]
browser = webdriver.Firefox()
url = browser.get('https://www.amazon.in/s?k=mouse+for+laptop&i=computers&rh=n%3A976392031%2Cn%3A1375248031%2Cn%3A1375412031%2Cn%3A1375420031&dc&qid=1600924627&rnid=976392031&ref=sr_nr_n_3')
for i in range(1,3):
    links=browser.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
    for link in links:
        href.append(link.get_attribute('href'))
    print('After {} iteration, number of links= {}'.format({i},{len(href)}))
    next_button=browser.find_element_by_xpath("//li[@class='a-last']//a")
    browser.get(next_button.get_attribute('href'))
    
data={}
data['Model']=[]
data['Price']=[]

for i in range(0, (len(href))):
    print("Product ", i+1, " in Progress!")
    try:
        
        browser.get(href[i])
        
        try:
            price=browser.find_element_by_xpath("//span[@id='priceblock_ourprice']").text
            
        except NoSuchElementException:
            print(i+1, ' Price not found. Moving On!!')
            continue
            
        try:
            title=browser.find_element_by_xpath("//span[@id='productTitle']").text
            
        except NoSuchElementException:
            print(i+1, ' Title not found. Moving On!!')
            continue
        
        price=price.replace(',', '')
        p=float(price.split(' ')[1])
        if p<=500.0:
            data['Model'].append(title)
            data['Price'].append(p)

    except StaleElementReferenceException:
        print(i+1, ' Stale Element!!')
        continue
```

```python
df=pd.DataFrame(data, columns=['Model', 'Price'])
# choosing only 20 products
df=df.iloc[:20,:]
df
```
